# Read HSI dataset
This sample notebook provides how to read each HSI dataset.

```python
import os
import glob
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
%%time
# Read all dataset
dataset = {}
for target in sorted(glob.glob("../data/*")):
    target_name = target.split('/')[-1]
    filenames = sorted(glob.glob(target + '/*'))
    dataset[target_name] = np.array([np.array(Image.open(f)) for f in filenames]).transpose(1, 2, 0)
    
print(dataset.keys())
```

Now, all the data are available as dictionary. For example, we show the average image for each data.

## 1. coffee

```python
print("Size of HSI: (height, width, spectral) = ", dataset["coffee"].shape)
mean_img = np.mean(dataset["coffee"], axis=2)
plt.imshow(mean_img)
```

```python
import matplotlib.patches as patches

# Create figure and axes
fig, ax = plt.subplots()

# Display the image
ax.imshow(mean_img)

# Create a Rectangle patch
rect = patches.Rectangle((130, 90), 20, 20, linewidth=1, edgecolor='r', facecolor='none')

# Add the patch to the Axes
ax.add_patch(rect)

plt.show()
```

```python
import matplotlib.patches as patches

# Create figure and axes
fig, ax = plt.subplots()

# Display the image
ax.imshow(dataset["coffee"][90:110, 130:150, 3])

# Add the patch to the Axes
#ax.add_patch(rect)

plt.show()
```

```python


extracted_dataset_1 = np.empty([400,96])
extracted_dataset_1.shape

itr = 0
for row_itr in range(90,110):
    for col_itr in range(130,150):

        extracted_dataset_1[itr,:] = dataset["coffee"][row_itr, col_itr, :]

        itr+=1
itr
```

```python
import copy
extracted_dataset_coffee = copy.deepcopy(extracted_dataset_1)
```

```python
extracted_dataset_coffee = ((extracted_dataset_coffee - extracted_dataset_coffee.min()) * (1/(extracted_dataset_coffee.max() - extracted_dataset_coffee.min()) * 255)).astype('uint8')
extracted_dataset_coffee
```
